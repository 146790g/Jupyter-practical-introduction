## 1. DataFrameを描画する

```python
import os
import pandas as pd
from bokeh.charts import Line, output_notebook, show

base_url='C:/Users/kohsuke maeda/TextBook-Jupyter-practical-introduction/anime'
anime_stock_returns=os.path.join(base_url,'anime_stock_returns.csv')

df=pd.read_csv(anime_stock_returns,index_col=0, parse_dates=['Date'])
df.head()
```

```python
p=Line(df,plot_width=800,plot_height=200)
show(p)
```

```python
p=Line(df,x='index',y='IG Port', plot_width=800, plot_height=200)
show(p)
```
